```python
import pandas as pd
df = pd.read_csv('Salary.csv')
df2 = pd.read_csv('Salary2.csv')
print(df)
inn_join = pd.merge(df, df2, on = "Age", how = "inner")
inn_left = pd.merge(df, df2, on = "Gender", how = "left")
print()
print(inn_join.head())
print()
print(inn_left.head())

```
