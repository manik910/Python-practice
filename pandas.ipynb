```python
import pandas as pd
date=(7,3,10,11,3,2,3)
d= pd.Series(date)
print(d)
print(d.sum())
print(d.max())
print(d.min())
print(d.mean())
print(d.median())
print(d.mode())
print(d.describe())
print(d.value_counts())
print(d.head())
print(d.tail())
```
