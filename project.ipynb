```python
import matplotlib.pyplot as plt
import pandas as pd
m = pd.read_csv("Salary.csv")

print(m.head())
print(m)

average_salary_by_gender = m.groupby('Gender')['Salary'].mean().reset_index()

plt.figure(figsize=(8, 6))
plt.bar(average_salary_by_gender['Gender'], average_salary_by_gender['Salary'], color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
print()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

m = pd.read_csv("Salary.csv")
print(m)

# Filter for Male employees
male_df = m[m['Gender'] == 'Male']


average_salary_female_by_experience = male_df.groupby('Years of Experience')['Salary'].mean().reset_index()

x = average_salary_male_by_experience['Years of Experience']
y = average_salary_male_by_experience['Salary']

plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='*', linestyle='-', color='purple')
plt.title('Average Salary for Males by Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Average Salary')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
