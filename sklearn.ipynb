```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

d = {
 "Company": ["Google", "Twitter", "Google", "LinkedIn"],
 "Role": ["Data Scientist", "Sales manager", "HR", "HR"]
}

df = pd.DataFrame(d)    #obj=class labelencoder()
label_encoder = LabelEncoder() #it  is method

df["Company_n"] = label_encoder.fit_transform(df['Company']) # fit_transform is usedto transform old company name to new name and alp to number
df["Role_n"] = label_encoder.fit_transform(df['Role'])

print()
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
d = {
      "Company": ["Google", "instgram", "gst", "LinkedIn"],
      "Role": ["Data Scientist", "Sales manager", "co", "HR"]
   }
df = pd.DataFrame(d)
label_encoder = LabelEncoder()
df['comp'] = label_encoder.fit_transform(df['Company'])
df['virat_k'] = label_encoder.fit_transform(df['Role'])
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

d = {
        "x" : [10, 20, 50, 40, 50],
        "y" : [75, 50, 75, 100, 125]
}

df = pd.DataFrame(d)
mm_scale = MinMaxScaler(feature_range = (0, 1))

print(df)

one_col = mm_scale.fit_transform(df[["x"]])
one_col2 = mm_scale.fit_transform(df[['y']])

print(one_col)
print(one_col2)
```

```python
import numpy as np
from sklearn.preprocessing import FunctionTransformer

a = [[20, 40], [30, 40], [70, 60]]

f = np.array(a)

def add_ten(x):       #adding 10 to a
 return x + 10

obj = FunctionTransformer(add_ten)

result = obj.transform(f)

print(f)
print()
print(result)
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
import pandas as pd

students = [
                   [85, 'M', 'verygood'],
                   [95, 'F', 'excellent'],
                   [60, None, 'good'],
                   [np.nan, 'M', 'average'],
                   [70, 'M', 'good'],
                   [np.nan, None, 'verygood'],
                   [60, 'F', 'verygood'],
                   [98, 'M', 'excellent']
]

cols = ['marks', 'gender', 'result']
df = pd.DataFrame(students, columns = cols)

print(df)

imputer = SimpleImputer(missing_values = np.nan, strategy =
'mean')

result = df['marks'].values.reshape(-1, 1)

df.marks = imputer.fit_transform(result)

print(df)
```
